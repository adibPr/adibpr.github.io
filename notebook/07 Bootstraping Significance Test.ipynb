```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = np.random.exponential(3, 1000)
sns.histplot(data)
```

```python
data_normal = np.random.normal(3, size=1000)
sns.histplot(data_normal)
```

```python
# turn on the edges: 
# https://stackoverflow.com/questions/43080259/no-outlines-on-bins-of-matplotlib-histograms-or-seaborn-distplots
plt.rcParams["patch.force_edgecolor"] = True
fig, ax = plt.subplots()
for a in [data, data_normal]:
    sns.histplot(a, ax=ax, bins=range(0, 10, 1), kde=True)
ax.set_xlim([0, 10])
```

```python
def bootstrap(data, tot_item=100, tot_iter=1000):
    mean = []
    for i in range(tot_iter):
        mean.append(np.random.choice(data, tot_item).mean())
    return mean

bs_exp = bootstrap(data)
bs_normal = bootstrap(data_normal)
```

```python
sns.histplot(bs_exp)
```

```python
sns.histplot(bs_normal)
```

```python
fig, ax = plt.subplots()
for a in [bs_exp, bs_normal]:
    sns.histplot(a, ax=ax, bins=[2+(0.10*i) for i in range(20)], kde=True)
# ax.set_xlim([0, 10])
```
